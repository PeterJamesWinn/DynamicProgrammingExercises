```python
'''Say that you're a traveller on a 2D grid. You begin in the top-left corner and your goal is to travel to the bottom-right corner. You may only move down or right. 

In how many ways can you travel to the goal on a grid with dimensions m * n?'''



def grid(m,n, memo = {}):  # grid traveller - how many ways to get from top right corner to bottom left corner of a grid ...
    if m <= 0 or n <= 0: raise Exception("indices too small to define a real grid"); return -1
    
    key = str(m) + "_"+ str(n)  # create key based on the two indices
    memo["1_1"] = 1  # set base case
    print(key)
    #print(memo.keys())

    if key in memo.keys(): # check if value for key already memoized
        print(key, " in memo.keys()")
        return memo[key]

    # value of a square is the value of the adjacent two squares summed, i.e. that of the square below and the square to the right
    #key_down = str(m-1) + "_"+ str(n)
    #key_right = str(m) + "_"+ str(n)
    if n -1 != 0: # checking if there is a square to the right and if so calculate it's value
        right_grid = grid(m, n-1)
    else: right_grid = 0 # if no square to the right there is nothing to add
    if m -1 != 0: # checking if square below the current one, that is on the grid
        down_grid = grid(m-1, n) 
    else: down_grid = 0  # if no square below set to zero.
    
    memo[key] = right_grid + down_grid
    return memo[key]




```

```python
#Testing grid traveller function grid()

assert grid(2,2) == 2, "failed grid(2,2)"
assert grid(4,4) == 20, "failed grid(4,4)"
assert grid(5,4) == 35, "failed grid(5,4)"
assert grid(18,18) == 2333606220, "failed grid(18,18)"
print("grid() passed all tests")
```
